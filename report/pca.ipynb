```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# load TF-IDF features
X = np.load("../data/processed/training_features.npy")
y = np.load("../data/processed/training_labels.npy")

# compute the PCA projection
pca = PCA(n_components=2, random_state=0)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(8,6))
scatter = plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap="viridis", alpha=0.6, s=10)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Projection of TF-IDF Feature Space (Linear Structure)")
plt.colorbar(scatter, label="Class")
plt.show()
```
